```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('NavBharat_all_pages.csv')
```

```python
df_new = pd.DataFrame(columns=df.columns)
```

```python
df_new
```

```python
df.shape
```

```python
s = {'@','#',':'}
```

```python
words_notneeded = {'कहानी','चंद्रमोहन शर्मा','चंद्रमोहन शर्मा','मूवी','नईफिल्म','नई फिल्म'}
```

```python
for i in range(65,65+26):
    s.add(chr(i))
for i in range(97,97+26):
    s.add(chr(i))
```

```python
for i in range(453,454):
    f=False
    cleaned = df.iloc[i].copy()
    original_review = cleaned['MovieReviews']
    new_review = ''
    cleaned['MovieReviews'] = ''
    for t in original_review:
        if(ord(t)==65533):
            print(f'{i}th review contains unkown character')
            f=True
            break
        if((t not in s)):
            new_review += t
    if(f==True):
        continue
    words = new_review.split()
    print(words)
    final_review = ' '.join([word for word in words if word not in words_notneeded])
    print(final_review)
    df_new = df_new.append(cleaned)
```

# Removing 0 ratings

```python
df = pd.read_csv('NavBharat_all_pages_cleaned.csv')
```

```python
df_new = pd.DataFrame(columns=df.columns)
```

```python
for i in range(df.shape[0]):
    if(df['MovieRatingCritic'].iloc[i][0]=='/'):
        continue
    df_new = df_new.append(df.iloc[i])
```

```python
df_new.to_csv('NavBharat_all_pages_cleaned_2.0.csv',index=False)
```

# Removing -1 rating

```python
df = pd.read_csv('Final_dataset.csv')
```

```python
df_new = pd.DataFrame(columns=df.columns)
```

```python
for i in range(df.shape[0]):
    if(df['Rating'].iloc[i]==-1):
        continue
    df_new = df_new.append(df.iloc[i].copy())
```

```python
df_new = df_new.drop(columns=['Unnamed: 3'])
```

```python
df_new.shape
```

```python
df_new = df_new.dropna(axis=0)
```

```python
df_new.shape
```

```python
df_new = df_new.sample(frac=1).reset_index(drop=True)
```

```python
df_new.to_csv('Final_dataset_2.0.csv',index=False)
```
